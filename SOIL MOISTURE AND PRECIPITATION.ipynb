### Accra
In Accra, the plot shows a clear seasonal pattern where precipitation peaks sharply around day 170-200, coinciding with the rainy season. The soil moisture then gradually decreases as the precipitation reduces, indicating the drying of the soil over time after the rainy season. The strong correlation between precipitation and soil moisture in Accra suggests that soil moisture responds directly to rainfall events.


### Kumasi
For Kumasi, precipitation is distributed more evenly throughout the year with several peaks. A significant increase in soil moisture corresponds to the first major rainfall peak around day 150-200, followed by a slow decline as rainfall decreases. The second peak in precipitation around day 250 is also reflected in a slight rebound in soil moisture. The relatively consistent fluctuation in both variables suggests that Kumasi experiences intermittent rainfall, with soil moisture being replenished during these events but not retaining as much moisture as seen in Accra.

### Sunyani
Sunyani shows a large peak in precipitation around day 150-250, during which soil moisture also increases significantly. After this peak period, both precipitation and soil moisture drop sharply, indicating a transition to drier conditions towards the end of the year. Compared to the other locations, Sunyani exhibits a more abrupt rise and fall in both precipitation and soil moisture, suggesting that the soil retains water less effectively after rainfall events.


### Wa
In Wa, the seasonal variation is more pronounced, with precipitation peaking around day 150-200 and soil moisture closely following this pattern. The steep increase and decline in both variables highlight the rainy season's short but intense nature. Soil moisture reaches its maximum during and just after the peak rainfall but decreases quickly as the rainfall subsides. The overall pattern indicates that Wa has a more arid climate with short, heavy rainfalls followed by dry periods, which is reflected in the sharp drop in soil moisture post-rainy season.

In summary, all locations show a strong relationship between precipitation and soil moisture, with the soil moisture levels closely following the rainfall patterns. However, the intensity and duration of these variables differ between locations, indicating regional differences in climate and soil characteristics.

```python
accra= pd.read_csv('monthly_accra.csv')
kumasi= pd.read_csv('monthly_kumasi.csv')
wa= pd.read_csv('monthly_wa.csv')
sunyani= pd.read_csv('monthly_sunyani.csv')
```

```python
locations = ['accra', 'kumasi', 'wa', 'sunyani']
data_paths = [f'monthly_{location}.csv' for location in locations]
data_list = [pd.read_csv(path) for path in data_paths]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

for ax, data, location in zip(axes.flatten(), data_list, locations):
    
    gwetroot_data = data[data['PARAMETER'] == 'GWETROOT']
    precipitation_data = data[data['PARAMETER'] == 'PRECTOTCORR_SUM']

    gwetroot_mean = gwetroot_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1).mean()
    precipitation_mean = precipitation_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1).mean()

    gwetroot_anomalies = gwetroot_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1) - gwetroot_mean
    precipitation_anomalies = precipitation_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1) - precipitation_mean

    ax.bar(gwetroot_anomalies.columns, gwetroot_anomalies.mean(), label='SOIL MOISTURE', alpha=0.7, color='red')

    ax2 = ax.twinx()
    ax2.bar(precipitation_anomalies.columns, precipitation_anomalies.mean(), label='PRECIPITATION', alpha=0.7, color='blue')

    ax.set_title(f'Anomalies for {location.capitalize()}')
    ax.set_xlabel('Month')
    ax.set_ylabel('Mean GWETROOT Anomaly')
    ax.set_xticks(range(12))
    ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
    ax.grid(True)
    ax.legend(loc='upper left')
    ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
locations = ['ACCRA', 'KUMASI', 'WA', 'SUNYANI']
data_paths = [f'monthly_{location.lower()}.csv' for location in locations]
data_list = [pd.read_csv(path) for path in data_paths]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

baseline_value = 0 

for ax, data, location in zip(axes.flatten(), data_list, locations):
    
    gwetroot_data = data[data['PARAMETER'] == 'GWETROOT']
    precipitation_data = data[data['PARAMETER'] == 'PRECTOTCORR_SUM']

    gwetroot_mean = gwetroot_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1).mean()
    precipitation_mean = precipitation_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1).mean()

    gwetroot_anomalies = gwetroot_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1) - gwetroot_mean
    precipitation_anomalies = precipitation_data.drop(['PARAMETER', 'YEAR', 'ANN'], axis=1) - precipitation_mean

    ax.bar(gwetroot_anomalies.columns, gwetroot_anomalies.mean(), label='SOIL MOISTURE', alpha=0.7, color='red')
    ax.axhline(y=baseline_value, color='black', linestyle='--', linewidth=0.5)  

    ax2 = ax.twinx()
    ax2.bar(precipitation_anomalies.columns, precipitation_anomalies.mean(), label='PRECIPITATION', alpha=0.7, color='blue')

    ax.set_title(f'ANOMALIES FOR {location}', fontweight='bold')
    ax.set_xlabel('MONTHS', fontweight='bold')
    ax.set_ylabel('MEAN GWETROOT ANOMALY', fontweight='bold')
    ax.set_xticks(range(12))
    ax.set_xticklabels(['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], rotation=45)
    ax.grid(True)
    ax.legend(loc='upper left')
    ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

These plots depict the **anomalies** in both **precipitation** and **soil moisture** (GWETROOT) for four locations: Accra, Kumasi, Wa, and Sunyani, over a series of months. Anomalies represent deviations from the long-term mean, so positive values indicate higher-than-average conditions, while negative values indicate lower-than-average conditions. 

###  **Accra**
- **Precipitation (blue bars)**: There is a strong negative precipitation anomaly in March, indicating significantly less rainfall than the long-term average. Precipitation anomalies turn positive in May and peak in July, suggesting above-average rainfall during these months, before declining again towards the end of the year.
- **Soil Moisture (red bars)**: Soil moisture follows a similar pattern, with a strong negative anomaly in March, suggesting drier-than-average conditions. It starts increasing from April through to July, correlating with the positive precipitation anomaly. This indicates that higher-than-average rainfall boosts soil moisture.

###  **Kumasi**
- **Precipitation**: There is a sharp positive precipitation anomaly in June, indicating an unusual increase in rainfall during this month. The rest of the months, especially after July, show a negative or near-zero anomaly, implying a drier period for the remaining months.
- **Soil Moisture**: The soil moisture anomaly mirrors the precipitation pattern. It shows a strong positive anomaly in April and June, where soil moisture increases significantly, likely due to higher rainfall. After July, the soil moisture anomaly becomes negative, suggesting the soil remains drier than usual during the later part of the year.

###  **Wa**
- **Precipitation**: The most notable anomaly in Wa occurs in July and August, where there is a strong positive precipitation anomaly. This means Wa received much more rainfall than the long-term average during these months. Precipitation anomalies in other months remain close to zero or slightly negative.
- **Soil Moisture**: Soil moisture anomalies are highly positive in April and July, likely reflecting the increase in precipitation during these months. There is a strong negative soil moisture anomaly towards the end of the year, especially in November and December, indicating much drier conditions than normal.

###  **Sunyani**
- **Precipitation**: There is a sharp positive anomaly in September, indicating higher-than-average rainfall. Other months show either slightly negative or near-zero anomalies, suggesting normal to drier-than-average conditions in those periods.
- **Soil Moisture**: The soil moisture anomaly is mostly negative for Sunyani throughout the year, with a strong negative anomaly in April and May, implying significantly drier-than-average conditions. Positive anomalies appear in September, likely due to the rainfall in that month, but the soil moisture quickly drops again afterwards.

### Summary
In all locations, there is a clear relationship between precipitation anomalies and soil moisture anomalies. Positive precipitation anomalies generally lead to increased soil moisture, while negative precipitation anomalies are associated with reduced soil moisture. The intensity and timing of these anomalies vary across the locations, reflecting the different climatic and environmental conditions in Accra, Kumasi, Wa, and Sunyani.
